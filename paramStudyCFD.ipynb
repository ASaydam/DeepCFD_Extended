```python
import matplotlib.pyplot as plt
from matplotlib import ticker, cm

def plotField(x,y,z, minv=None, maxv=None, titleLabel="Velocity Plot"):
  if minv is None:
    minv = np.min(z)
  if maxv is None:
    maxv = np.max(z)
  fig = plt.figure()
  fig = plt.gcf()
  fig.set_size_inches(20,5)    
  ax = fig.add_subplot(1, 1, 1)
  #plt.tripcolor(x,y,z)
  plt.title(titleLabel)
  plt.xlabel("x")
  plt.ylabel("y")
  cs = ax.tricontourf(x,y,z, levels=np.linspace(minv, maxv), cmap=cm.rainbow)
  cbar = fig.colorbar(cs)
  plt.show()
  plt.close()
```

```python
time = 1399
x = np.nan_to_num(data250[0,time,:])
y = np.nan_to_num(data250[1,time,:])
z = np.nan_to_num(data250[2,time,:])

plotField(x,y,z)
```
